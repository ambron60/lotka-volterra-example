```python
from scipy import integrate
import pylab as p
```

```python
# Definition of parameters
a = 1.
b = .25
c = 0.2
d = 0.6
```

```python
def dX_dt(X, t=0):
    """ Return the growth rate of fox and rabbit populations. """
    return array([a*X[0]-b*X[0]*X[1], c*X[0]*X[1]-d*X[1]])
```

```python
# Equilibrium occurs when the growth rate is equal to 0. This gives two fixed points:
X_f0 = array([0., 0.])
X_f1 = array([c/(d*b), a/b])
```

```python
# Integrate the ODE
t = linspace(0, 50, 1000) # time
X0 = [10, 6] # initial fox and rabbit populations
X = integrate.odeint(dX_dt, X0, t)
```

```python
# Plot the evolution of both populations:
rabbits, foxes = X.T
f1 = p.figure()
p.plot(t, rabbits, 'b-', label='Rabbits')
p.plot(t, foxes  , 'r-', label='Foxes')
p.grid()
p.legend(loc='best')
p.xlabel('time')
p.ylabel('population')
p.title('Evolution of rabbit and fox populations')
```
